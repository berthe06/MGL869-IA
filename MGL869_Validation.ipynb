```python
#Define the data download location and the location to save TFRecord files
import os
import opendatasets as od

# Define the data download location and the location to save TFRecord files
params = {}
params['download_data_location'] = '/content/intel-image-classification'
params['tfrecord_location'] = '/content/tfrecords/'

# Create directories if they don't exist
os.makedirs(params['download_data_location'], exist_ok=True)
os.makedirs(params['tfrecord_location'], exist_ok=True)

# Download the dataset
dataset_url = 'https://www.kaggle.com/datasets/puneet6060/intel-image-classification'
od.download(dataset_url, params['download_data_location'])
```

# **Génération du schéma**

```python
!pip install tfx
!pip install tensorflow_data_validation
```

```python
import tensorflow_data_validation as tfdv
from tensorflow_metadata.proto.v0 import schema_pb2

# Path to your TFRecord file
tfrecord_path = '/content/tfrecords/train.tfrecord'

# Generate statistics from the TFRecord file
stats = tfdv.generate_statistics_from_tfrecord(data_location=tfrecord_path)

# Infer the schema from the statistics
schema = tfdv.infer_schema(stats)

# Define the width feature
width_feature = schema_pb2.Feature()
width_feature.name = 'width'
width_feature.type = schema_pb2.INT
width_feature.int_domain.min = 150
width_feature.int_domain.max = 150

# Define the height feature
height_feature = schema_pb2.Feature()
height_feature.name = 'height'
height_feature.type = schema_pb2.INT
height_feature.int_domain.min = 150
height_feature.int_domain.max = 150

# Add the features to the schema
schema.feature.add().CopyFrom(width_feature)
schema.feature.add().CopyFrom(height_feature)

# Display the modified schema
tfdv.display_schema(schema)
```

# **Valider les données avec taille des images**

```python
# Validate the evaluation data against the schema
anomalies = tfdv.validate_statistics(statistics=stats, schema=schema)

# Display anomalies
tfdv.display_anomalies(anomalies)
```
